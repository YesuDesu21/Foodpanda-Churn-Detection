# Data Inspection

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle('../data/processed/train.pkl')
```

```python
df.head(5)
```

# Feature Aggregation

```python
df['order_timespan_days'] = (df['last_order_date'] - df['order_date']).dt.days
```

```python
df['order_quantity'] = (df['quantity'] * df['order_frequency'])
```

# Unused feature removal

# df1

```python
df1 = df[['rating','delivery_status','churned']]
```

```python
delivery_dummies = pd.get_dummies(df1['delivery_status'], dtype=int)

df_encoded = df1.join(delivery_dummies)

df_encoded = df_encoded.drop('delivery_status', axis=1)
```

```python
df_encoded.shape
```

# df2 

```python
#Creating another df with different features/added features
df2 = df[['restaurant_name','churned','quantity','loyalty_points','city','age','gender','order_frequency','price','rating']]
```

```python
restaurant_dummies = pd.get_dummies(df2['restaurant_name'], dtype=int)
city_dummies = pd.get_dummies(df2['city'], dtype=int)
gender_dummies = pd.get_dummies(df2['gender'], dtype=int)
 
df2_encoded = df2.join(restaurant_dummies).join(city_dummies).join(gender_dummies)
df2_encoded = df2_encoded.drop(['restaurant_name','city','gender'], axis=1)
```
